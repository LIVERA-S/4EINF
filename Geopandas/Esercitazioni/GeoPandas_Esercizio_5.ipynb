```python
# caricare il geodataset dei quartieri di Milano ds964_nil_wm.zip
quartieri = geopandas.read_file("/content/drive/MyDrive/Colab Notebooks/4E INF/GEOpands/quartieri di Milano/ds964_nil_wm.zip")
quartieri
```

```python
# installare (se necessario) e importare contextily e visualizzare la mappa di Milano con l'indicazione dei quartieri in nero (solo confini)
# e le piste ciclabili
ax=quartieri.to_crs(epsg=3857).plot(figsize=(18,15),facecolor="none",edgecolor="K", linewidth=5)
ax = bike_ciclabili.to_crs(epsg=3857).plot(ax=ax, linewidth=5)
contextily.add_basemap(ax)
```

```python
# trovare il quartiere meno esteso
min = quartieri['Shape_Area'].min()
meno_est = quartieri[quartieri['Shape_Area'] == min]
meno_est 
```

```python
# trovare il quartiere piÃ¹ esteso
max = quartieri['Shape_Area'].max()
piu_est = quartieri[quartieri['Shape_Area'] == max]
piu_est 
```

```python
# visualizzare i due quartieri sulla mappa di Milano con i colori indicati nella figura sotto
ax= piu_est.to_crs(epsg=3857).plot(figsize=(15,15),facecolor="yellow", edgecolor='k', alpha=0.5, linewidth=2)
ax = meno_est.to_crs(epsg=3857).plot(ax=ax, linewidth=5,facecolor="red", alpha=0.5)
contextily.add_basemap(ax)
```

```python
# salvare in un geodataframe le informazioni sul quartiere di Affori, prendendole dal geodataframe dei quartieri
Affori = quartieri.loc[quartieri['NIL']=='AFFORI']
Affori
```

```python
Affori_sq = Affori.geometry.squeeze()
Affori_sq
```

```python
# trovare le piste ciclabili che passano per il quartiere di Affori
piste_ciclabili_affori = bike_ciclabili[bike_ciclabili.intersects(Affori_sq)]
piste_ciclabili_affori
```

```python
# visualizzare su una mappa il quartiere di Affori e le piste ciclabili che lo attraversano
# evidenziare bene in rosso le piste ciclabili come nell'esempio sotto
```

```python
ax = piste_ciclabili_affori.to_crs(epsg=3857).plot(figsize=(18,15), linewidth=4, facecolor='none', edgecolor='red')
contextily.add_basemap(ax)
```
