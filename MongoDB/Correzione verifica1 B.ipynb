6.il numero di impianti per ogni gestore

```python
result = db.impianti.aggregate([{ "$group" : {"_id": "$gestore", "total" : {"$sum" : 1}}}])
df2 = pd.DataFrame(list(result))
df2
```

7. i gestori che hanno piu di 100, ordinare in ordine decrescente a partire dal gestore ce ha piu impianti

```python
result = db.impianti.aggregate([{ "$group" : {"_id": "$gestore", "total" : {"$sum" : 1}}},
                                { "$match": { "total": { "$gt": 100 }}},
                                { "$sort": {"total": -1 }}])
df2 = pd.DataFrame(list(result))
df2
```
