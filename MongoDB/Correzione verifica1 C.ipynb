**La collection contiene informazioni sulle vendite di uno store online che ha anche negozi fisici.**

**Analizzare con attenzione le informazioni riportate e realizzare poi un notebook su Colab che permetta di estrarre le seguenti informazioni:**


**1.le mail degli utenti che hanno usato i coupon per un acquisto**

```python
result=db.sales.find({"couponUsed":True},{"mail":"$customer.email","_id":0})
df=pd.DataFrame(list(result))
df
```

```python
result=db.sales.find({"couponUsed":True},{"customer.email":1,"_id":0})
df=pd.DataFrame(list(result))
df
```

**2.le mail degli utenti che hanno comperato articoli per la scuola**


```python
result=db.sales.find({"items.tags":{"$regex":"school"}},{"customer.email":1,"_id":0})
df=pd.DataFrame(list(result))
df
```

**3.gli ordini effettuati nel marzo del 2013**

```python
import datetime
result=db.sales.find({"$and":[{"saleDate":{"$gte":datetime.datetime(2013,3,1)}},{"saleDate":{"$lte":datetime.datetime(2013,3,31,23,59)}}]})
df=pd.DataFrame(list(result))
df
```

**4.l’età media dei clienti** 


```python
result=db.sales.aggregate([
                           {"$group":{"_id":"|>","etaMedia":{"$avg":"$customer.age"}}}
])
df=pd.DataFrame(list(result))
df
```

**5.il numero di clienti uomini e donne**

```python
result=db.sales.aggregate([
                           {"$group":{"_id":"$customer.gender","total":{"$sum":1}}}
])
df=pd.DataFrame(list(result))
df
```

**6.il numero di acquisti per ogni categoria. Ordinare le categorie in ordine decrescente**

```python
result=db.sales.aggregate([
                           {"$unwind":"$items"},
                           {"$unwind":"$items.tags"},
                           {"$group":{"_id":"$items.tags","total":{"$sum":1}}},
                           {"$sort":{"total":-1}}                                                 
])
df=pd.DataFrame(list(result))
df
```

**7.il numero di acquisti per ogni negozio (‘storeLocation’)**

```python
result=db.sales.aggregate([
                           {"$group":{"_id":"$storeLocation","total":{"$sum":1}}}                                              
])
df=pd.DataFrame(list(result))
df
```
