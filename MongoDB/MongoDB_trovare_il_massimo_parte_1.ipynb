```python
#fare un $group per trovare il valore massimo sul campo customer.age, utilizzxando il push inseriamo i dati all'interno di un nuovo campoo che si chiama  customerdata
result = db.vendite.aggregate([{ "$group": {"_id": "maxAge","maxAge": {"$max": "$customer.age"},
        "customerdata": {"$push": {"age": "$customer.age", "email": "$customer.email"}}}}])
df = pd.DataFrame(list(result))
df
```

```python
#fare un $unwind per estrasse i singoli cutomer dalla lista customerdata
result = db.vendite.aggregate([{ "$group": {"_id": "maxAge","maxAge": {"$max": "$customer.age"},
        "customerdata": {"$push": {"age": "$customer.age", "email": "$customer.email"}}}},{'$unwind': '$customerdata'}])
df = pd.DataFrame(list(result))
df
```

```python
#piu vechchi clienti (fare un confronto tra max age e customerdata.age per vedere sono uguali)
# si usa l'operatore $match. 
```

```python
result = db.vendite.aggregate([
     
    { "$group": {
        "_id": "maxAge",
        "maxAge": {
            "$max": "$customer.age"
        },
        "customerdata": {
            "$push": {
                "age": "$customer.age",
                "email": "$customer.email"
            }
        }}
       },
       
      {'$unwind': '$customerdata'
    },
    {'$match':{'$expr':{"$eq":["$maxAge", "$customerdata.age"]}}}
    ])
df = pd.DataFrame(list(result))
df
```
